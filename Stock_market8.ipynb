```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool, ColumnDataSource

start=datetime.datetime(2016,3,1)
end=datetime.datetime(2016,3,10)
cds=ColumnDataSource(df)

df=data.DataReader(name="GOOG",data_source="iex",start=start,end=end)

df["Middle"]=(df["open"]+df["close"])/2
df["Height"]=abs(df["close"]-df["open"])
df.index=[df["close"] > df["open"]]
#we are trying to get only the dates on which the closing value was higher than opening value

#date_increase=df.index[df["close"] > df["open"]]
#date_decrease=df.index[df["close"] < df["open"]]

def inc_dec(c, o):
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value


df["Status"]=[inc_dec(c,o) for c, o in zip(df.close,df.open)] #c for closing value and o for opening value



p=figure(x_axis_type='datetime', width=1000, height=300, sizing_mode='scale_width', title="Motion Graph")
#p.title.text="Candlestick Chart"
p.grid.grid_line_alpha=0.3

hours_12=12*60*60*1000  #width of the rectangle


#abs(df.open-df.close) gives absolute value of open-close in case close>open, this will act as height of rectangle
#p.quad
#p.quad
#in rectangles we pass 4 values, one for each corner (x-coord, y-coord, width, height)
p.segment(df.index, df.high, df.index, df.low, color="Black")


p.rect("Status", "Middle",
       hours_12, "Height",fill_color="#CCFFFF",line_color="black", source=cds)
p.rect("Status","Middle",
       hours_12, "Height",fill_color="#FF3333",line_color="black", source=cds)


output_file("cs.html")

#p.rect(left="Start",right="End",bottom=0,top=1,color="green",source=cds)


show(p)
```
